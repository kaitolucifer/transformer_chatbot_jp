### Transformer Position Encoding

transformerの完全構築はこの[ノートブック](https://github.com/kaitolucifer/transformer_chatbot_jp/blob/master/transformer_chatbot_jp.ipynb)をご参照ください。

transformerアーキテクチャはCNNやRNNではなく、スタックされたattention layerを利用する。これによって、長期的な依存関係は簡単に学習できるが、シーケンスの中のアイテムの相対位置に関する情報はビルドインされていない。

モデルがこの情報にアクセスできるよう、transformerは入力にposition encodingを足すようにしている。

encodingはそれぞれの位置のsinとcosからなるベクトルで、全てのsin-cosペアは違う頻度で回転する。

近接場所は似たposition-encodingベクトルを持っている。

### Imports

```
import numpy as np
import matplotlib.pyplot as plt
```

ベクトルの深さに渡って、角速度の範囲は`1 [rads/step]`から`min_rate [rads/step]`

角速度の式は：
$$angle_rate_d=(min_rate)^{d/d_{max}}$$

```
num_positions = 50
depth = 512
min_rate = 1/10000

assert depth%2 == 0, "Depth must be even."
angle_rate_exponents = np.linspace(0,1,depth//2)
angle_rates = min_rate**(angle_rate_exponents)
```

指数項（exponent）は0から1で、`angle_rates`を1から`min_rate`へ指数的に減らしていく。

```
plt.semilogy(angle_rates)
plt.xlabel('Depth')
plt.ylabel('Angle rate [rads/step]')
```

angle ratesとpositionsブロードキャスト掛け算はposition encoding anglesをdepthにマッピングする関数になる。

```
positions = np.arange(num_positions) 
angle_rads = (positions[:, np.newaxis])*angle_rates[np.newaxis, :]
```

```
plt.figure(figsize = (14,8))
plt.pcolormesh(
    # ラジアンを度（degree）に変換する、そして360以内に収める
    angle_rads*180/(2*np.pi) % 360,
    # 周期性のあるcolormapそ使って、color(0) == color(360)にする
    cmap='hsv', vmin=0, vmax=360)

plt.xlim([0,len(angle_rates)])
plt.ylabel('Position')
plt.xlabel('Depth')
bar = plt.colorbar(label='Angle [deg]')
bar.set_ticks(np.linspace(0,360,6+1))
```

角度はモデルの入力としてよくない（無制限で、日連続だから）。なので、sinとcosを計算する：

```
sines = np.sin(angle_rads)
cosines = np.cos(angle_rads)
pos_encoding = np.concatenate([sines, cosines], axis=-1)
```

```
plt.figure(figsize=(14,8))
plt.pcolormesh(pos_encoding, 
               # 分離系のcolormapを使って0の位置をはっきりさせる
               cmap='RdBu', vmin=-1, vmax=1)
plt.xlim([0,depth])
plt.ylabel('Position')
plt.xlabel('Depth')
plt.colorbar()
```

### Nearby positions
隣接位置は似たposition-encodingベクトルを持つ。
これを証明するため、position 20のposition encodingを他の全ての位置と比較する。

```
pos_encoding_at_20 = pos_encoding[20]

dots = np.dot(pos_encoding,pos_encoding_at_20)
SSE = np.sum((pos_encoding - pos_encoding_at_20)**2, axis=1)
```

どの手法で比較しても、20の位置で一番似ていて、遠くなれなるほど分離する。

```
plt.figure(figsize=(10,8))
plt.subplot(2,1,1)
plt.plot(dots)
plt.ylabel('Dot product')
plt.subplot(2,1,2)
plt.plot(SSE)
plt.ylabel('SSE')
plt.xlabel('Position')
```

### Relative positions
[論文](https://arxiv.org/pdf/1706.03762.pdf)の3.5節に示されたように、相対 position encodingは現在位置の線形関数に書き変えることができる。

これを証明するため、このセッションは行列を作って、その相対 position encodingを計算する。

```
def transition_matrix(position_delta, angle_rates = angle_rates):
    # 行列掛け算として実装される:
    #    sin(a+b) = sin(a)*cos(b)+cos(a)*sin(b)
    #    cos(a+b) = cos(a)*cos(b)-sin(a)*sin(b)

    # b
    angle_delta = position_delta*angle_rates

    # sin(b), cos(b)
    sin_delta = np.sin(angle_delta)
    cos_delta = np.cos(angle_delta)

    I = np.eye(len(angle_rates))

    # sin(a+b) = sin(a)*cos(b)+cos(a)*sin(b)
    update_sin = np.concatenate([I*cos_delta, I*sin_delta], axis=0)

    # cos(a+b) = cos(a)*cos(b)-sin(a)*sin(b)
    update_cos = np.concatenate([-I*sin_delta, I*cos_delta], axis=0)

    return np.concatenate([update_sin, update_cos], axis=-1)
```

例えば、現在のposition encodingから10歩前のpositionを計算する行列を作る

```
position_delta = -10
update = transition_matrix(position_delta)
```

この行列を全てのposition encodingベクトルに適用すれば、全てのposition encoding vectorは-10歩後退し、position-enodingのシフトマッピングが実現される。

```
plt.figure(figsize=(14,8))
plt.pcolormesh(np.dot(pos_encoding,update), cmap='RdBu', vmin=-1, vmax=1)
plt.xlim([0,depth])
plt.ylabel('Position')
plt.xlabel('Depth')
```

数値精確的には正しい結果。

```
errors = np.dot(pos_encoding,update)[10:] - pos_encoding[:-10]
abs(errors).max()
```
